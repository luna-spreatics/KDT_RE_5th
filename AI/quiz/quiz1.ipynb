## PyTorch Quiz 1 : Tensor 초기화

````python
import torch
````

### 1. 빈 Tensor 생성하기
다음 조건에 맞는 Tensor를 생성하세요.
- 3x4 크기의 Tensor
- 값이 초기화되지 않은 상태

````python
# 여기에 코드를 작성합니다
t1 = torch.empty(3,4)
print(t1)
````

### 2. 0으로 채워진 Tensor 생성하기
다음과 같은 Tensor를 생성하세요.
- 2x5x3 크기의 Tensor
- 데이터 타입은 float64

````python
# 여기에 코드를 작성합니다
t2 = torch.zeros(2,5,3, dtype=torch.float64)
print(t2)
````

### 3. 1로 채워진 Tensor 생성하기
다음 조건에 맞는 Tensor를 생성하세요.
- 4x4 크기의 Tensor
- 데이터 타입은 int

````python
# 여기에 코드를 작성합니다
t3 = torch.ones(4,4, dtype=torch.int)
print(t3)
````

### 4. 랜덤 값으로 채워진 Tensor 생성하기
다음 조건에 맞는 Tensor를 생성하세요.
- 3x3 크기의 Tensor
- 모든 값이 0과 1 사이의 랜덤 값

````python
# 여기에 코드를 작성합니다
t4 = torch.rand(3,3)
print(t4)
````

### 5. 다차원 Tensor 생성하기
다음과 같은 다차원 Tensor를 생성하세요.
- 값이 다음과 같은 2x3의 2D 배열
    ```
    [[1, 2, 3],
     [4, 5, 6]]
    ```

````python
# 여기에 코드를 작성합니다
t5 = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])
print(t5)
````

### 6. Tensor의 값을 변경하기
1. 표준정규분포의 값을 랜덤으로 갖는 2x2 Tensor를 생성하세요.
2. 생성된 Tensor의 속성을 물려받은 원소가 0인 새로운 텐서를 만들어보세요. 

````python
# 여기에 코드를 작성합니다
t6 = torch.randn(2,2, dtype=torch.float64)
print(t6)
t7 = torch.zeros_like(t6)
print(t7)
````
