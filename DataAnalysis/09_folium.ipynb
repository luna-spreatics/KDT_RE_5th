## 09. Folium
- 지도 시각화 라이브러리

```python
import folium
```

```python
# 기본 지도 생성
m = folium.Map(location=[37.61123411337214,126.91731045710175], zoom_start=18, )

# 기본 마커
folium.Marker([37.61123411337164,126.91731045719163], 
              popup="Subway",
              tooltip="구산역",
              icon=folium.Icon(color="black", icon="fa-solid fa-bus",icon_color="red", prefix="fa-solid")
              ).add_to(m)


m
```
